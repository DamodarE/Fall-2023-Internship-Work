a.) Convolution is the expression of the input and output of a linear time invariant (LTI) system by combinging two signals to form a third signal. The two types of convolution are linear and circular. Linear convolution is for aperiodic signals while circular signals are for periodic signals that repeat. Convolution helps map the effects of the LTI system onto the impulse signal. 

Mathematical Definitions: 

Time Domain:
$$y(n) \sum_{k = -\infty}^{\infty} x(k)h(n-k) $$

Frequency Domain:
$$F[x(t)y(t)] = \frac{1}{2\pi} X(\omega) * Y(\omega)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal


fig,ax = plt.subplots(4,3,  figsize=(20, 20))

cols = ['Signal {}'.format(col) for col in range(1, 4)]
for a, col in zip(ax[0], cols):
    a.set_title(col)


for i in range(4):
    for j in range(3):
        ax[i, j].grid()
        ax[i, j].set_ylabel('Amplitude')
        ax[i, j].set_xlabel('Time')


t = np.linspace(0, 1, 1000)
sig1 = signal.square(2 * np.pi * 5 * t)


t = np.linspace(0, 1, 500)
sig2 = signal.sawtooth(2 * np.pi * 5 * t, 0.5)

plt.subplot(4,3,1)
plt.axhline(y=0, color='k')
plt.xlim(0,500)
plt.ylim(-1.01,1.01)
plt.plot(sig1)


plt.subplot(4,3,2)
plt.axhline(y=0, color='k')
plt.xlim(0,500)
plt.ylim(-1.01,1.01)
plt.plot(sig1)
plt.plot(0)


plt.subplot(4,3,3)
plt.axhline(y=0, color='k')
conv = np.convolve(sig1, sig1, mode='valid')
plt.xlim(0,500)
plt.ylim(-25,25)
plt.plot(conv)

plt.subplot(4,3,4)
plt.axhline(y=0, color='k')
plt.xlim(0,500)
plt.ylim(-1.01,1.01)
plt.plot(sig1)

plt.subplot(4,3,5)
plt.axhline(y=0, color='k')
plt.xlim(0,500)
plt.ylim(-1.01,1.01)
plt.plot(sig2)

plt.subplot(4,3,6)
plt.axhline(y=0, color='k')
conv = np.convolve(sig1, sig2, mode='valid')
plt.xlim(0,500)
plt.ylim(-25,25)
plt.plot(conv)

plt.subplot(4,3,7)
plt.axhline(y=0, color='k')
plt.xlim(0,500)
plt.ylim(-1.01,1.01)
plt.plot(sig2)

plt.subplot(4,3,8)
plt.axhline(y=0, color='k')
plt.xlim(0,500)
plt.ylim(-1.01,1.01)
plt.plot(sig1)

plt.subplot(4,3,9)
plt.axhline(y=0, color='k')
conv = np.convolve(sig2, sig1, mode='valid')
plt.xlim(0,500)
plt.ylim(-25,25)
plt.plot(conv)

plt.subplot(4,3,10)
plt.axhline(y=0, color='k')
plt.xlim(0,500)
plt.ylim(-1.01,1.01)
plt.plot(sig2)

plt.subplot(4,3,11)
plt.axhline(y=0, color='k')
plt.xlim(0,500)
plt.ylim(-1.01,1.01)
plt.plot(sig2)

plt.subplot(4,3,12)
plt.axhline(y=0, color='k')
conv = np.convolve(sig2, sig2, mode='valid')
plt.xlim(0,500)
plt.ylim(-25,25)
plt.plot(conv)


plt.show()
```
